```python
# @title
from dataclasses import dataclass
from typing import List, Tuple, Dict, Optional
import plotly.graph_objects as go

@dataclass
class VPhase:
    """Clase que representa una fase del modelo en V con estilo profesional."""
    name: str
    description: str
    color: str
    coordinates: Tuple[float, float]
    width: float = 2.4            # Ancho ampliado para mayor legibilidad
    height: float = 1.6           # Alto ampliado para mayor legibilidad
    text_color: str = 'white'     # Color de texto predeterminado
    title_font_size: int = 22     # Tama√±o de fuente para el t√≠tulo
    desc_font_size: int = 16      # Tama√±o de fuente para la descripci√≥n
    hover_text: str = None        # Texto adicional para el hover (opcional)

class EnhancedVModelDiagram:
    """Generador de diagramas de Modelo en V con dise√±o profesional y alta legibilidad."""

    def __init__(
        self,
        title: str = "MODELO EN V",
        subtitle: str = "Desarrollo y Validaci√≥n de Software",
        company: str = "EUROLOGISTIC",
        copyright_text: str = "¬© EUROLOGISTIC - Metodolog√≠a de Desarrollo de Software",
        theme: str = "corporate"   # Opciones: 'corporate', 'modern', 'classic'
    ):
        """
        Inicializa el generador del diagrama del Modelo en V.

        Args:
            title: T√≠tulo principal del diagrama
            subtitle: Subt√≠tulo descriptivo
            company: Nombre de la empresa para mostrar en el t√≠tulo
            copyright_text: Texto de copyright a mostrar
            theme: Tema de colores ('corporate', 'modern', 'classic')
        """
        self.title = title
        self.subtitle = subtitle
        self.company = company
        self.copyright_text = copyright_text
        self.theme = theme

        # Inicializar figura
        self.fig = go.Figure()

        # Configurar tema y colores
        self._set_theme(theme)

        # Crear fases del modelo
        self.left_phases = self._create_development_phases()
        self.central_phase = self._create_implementation_phase()
        self.right_phases = self._create_validation_phases()

        # Configurar layout base
        self._configure_layout()

    def _set_theme(self, theme: str):
        """Configura las paletas de colores y estilos seg√∫n el tema seleccionado."""
        themes = {
            'corporate': {
                'primary': '#003366',
                'secondary': '#004D99',
                'accent': '#FF8C00',
                'development': ['#0A3D7E', '#1C5CAA', '#2F7BD6'],
                'validation': ['#2F7BD6', '#1C5CAA', '#0A3D7E'],
                'line': '#AAAAAA',
                'connection_line': '#DDDDDD',
                'bg': 'white',
                'title_color': '#003366',
                'text_color': 'white',
                'desc_color_outside': '#333333',
                'font_family': 'Arial, sans-serif'
            },
            'modern': {
                'primary': '#2C3E50',
                'secondary': '#34495E',
                'accent': '#E74C3C',
                'development': ['#2C3E50', '#34495E', '#3A536B'],
                'validation': ['#3A536B', '#34495E', '#2C3E50'],
                'line': '#95A5A6',
                'connection_line': '#BDC3C7',
                'bg': 'white',
                'title_color': '#2C3E50',
                'text_color': 'white',
                'desc_color_outside': '#34495E',
                'font_family': 'Helvetica, Arial, sans-serif'
            },
            'classic': {
                'primary': '#1B4F72',
                'secondary': '#21618C',
                'accent': '#D35400',
                'development': ['#1B4F72', '#2874A6', '#3498DB'],
                'validation': ['#3498DB', '#2874A6', '#1B4F72'],
                'line': '#85929E',
                'connection_line': '#D5D8DC',
                'bg': 'white',
                'title_color': '#1B4F72',
                'text_color': 'white',
                'desc_color_outside': '#273746',
                'font_family': 'Times New Roman, serif'
            }
        }

        # Aplicar tema seleccionado o tema corporativo por defecto
        self.colors = themes.get(theme, themes['corporate'])

        # Configuraci√≥n de dise√±o
        self.design = {
            'width': 1400,               # Ancho ampliado para mejor espaciado
            'height': 900,               # Alto amplio para mejor visualizaci√≥n
            'margins': {'t': 120, 'b': 100, 'l': 80, 'r': 80},
            'title_size': 28,
            'subtitle_size': 18,
            'annotation_size': 20,
            'copyright_size': 14,
            'line_width': 1.5,
            'connection_line_width': 1,
            'phase_border_width': 1.5,
            'phase_title_font_size': 22,
            'phase_desc_font_size': 16,
            'desc_outside_font_size': 14,
            'hover_font_size': 14
        }

    def _create_development_phases(self) -> List[VPhase]:
        """Crea las fases del lado izquierdo (Desarrollo)."""
        return [
            VPhase(
                name="Requisitos",
                description="Definici√≥n y validaci√≥n de requerimientos",
                color=self.colors['development'][0],
                coordinates=(1, 6),
                hover_text="‚Ä¢ Captura de requisitos<br>‚Ä¢ Definici√≥n de alcance<br>‚Ä¢ An√°lisis de viabilidad"
            ),
            VPhase(
                name="An√°lisis",
                description="Modelado del sistema y arquitectura",
                color=self.colors['development'][1],
                coordinates=(3, 4),
                hover_text="‚Ä¢ An√°lisis funcional<br>‚Ä¢ Modelado de procesos<br>‚Ä¢ Especificaciones t√©cnicas"
            ),
            VPhase(
                name="Dise√±o",
                description="Dise√±o detallado de componentes",
                color=self.colors['development'][2],
                coordinates=(5, 2),
                hover_text="‚Ä¢ Dise√±o de arquitectura<br>‚Ä¢ Dise√±o de interfaces<br>‚Ä¢ Dise√±o de bases de datos"
            )
        ]

    def _create_implementation_phase(self) -> VPhase:
        """Crea la fase central (Implementaci√≥n)."""
        return VPhase(
            name="Implementaci√≥n",
            description="Codificaci√≥n y construcci√≥n",
            color=self.colors['accent'],
            coordinates=(7, 0),
            width=2.8,                # M√°s ancho para destacar
            height=1.8,               # M√°s alto para destacar
            title_font_size=24,       # Fuente m√°s grande para destacar
            desc_font_size=18,
            hover_text="‚Ä¢ Desarrollo de c√≥digo<br>‚Ä¢ Pruebas unitarias<br>‚Ä¢ Revisi√≥n de c√≥digo<br>‚Ä¢ Control de calidad"
        )

    def _create_validation_phases(self) -> List[VPhase]:
        """Crea las fases del lado derecho (Validaci√≥n)."""
        return [
            VPhase(
                name="Verificaci√≥n",
                description="Verificaci√≥n contra requisitos",
                color=self.colors['validation'][0],
                coordinates=(9, 2),
                hover_text="‚Ä¢ Revisi√≥n t√©cnica<br>‚Ä¢ Verificaci√≥n vs requisitos<br>‚Ä¢ Pruebas de componentes"
            ),
            VPhase(
                name="Integraci√≥n",
                description="Integraci√≥n y pruebas de m√≥dulos",
                color=self.colors['validation'][1],
                coordinates=(11, 4),
                hover_text="‚Ä¢ Pruebas de integraci√≥n<br>‚Ä¢ Validaci√≥n de interfaces<br>‚Ä¢ Pruebas de subsistemas"
            ),
            VPhase(
                name="Validaci√≥n",
                description="Validaci√≥n completa del sistema",
                color=self.colors['validation'][2],
                coordinates=(13, 6),
                hover_text="‚Ä¢ Pruebas de aceptaci√≥n<br>‚Ä¢ Validaci√≥n de requisitos<br>‚Ä¢ Auditor√≠a de calidad"
            )
        ]

    def _configure_layout(self):
        """Configura el layout base del diagrama."""
        title_text = f"<b>{self.title}</b><br><span style='font-size:{self.design['subtitle_size']}px'>{self.subtitle}</span>"
        if self.company:
            title_text += f"<br><span style='font-size:{self.design['subtitle_size']}px'>{self.company}</span>"

        self.fig.update_layout(
            title={
                'text': title_text,
                'x': 0.5,
                'y': 0.98,
                'xanchor': 'center',
                'yanchor': 'top',
                'font': {
                    'size': self.design['title_size'],
                    'family': self.colors['font_family'],
                    'color': self.colors['title_color']
                }
            },
            xaxis={
                'visible': False,
                'range': [-0.5, 14.5]
            },
            yaxis={
                'visible': False,
                'range': [-1.5, 7.5]
            },
            plot_bgcolor=self.colors['bg'],
            paper_bgcolor=self.colors['bg'],
            width=self.design['width'],
            height=self.design['height'],
            margin=self.design['margins'],
            showlegend=False,
            annotations=[
                {
                    'text': self.copyright_text,
                    'showarrow': False,
                    'xref': "paper",
                    'yref': "paper",
                    'x': 0.5,
                    'y': -0.08,
                    'font': {
                        'size': self.design['copyright_size'],
                        'color': 'gray',
                        'family': self.colors['font_family']
                    }
                }
            ]
        )

        # Agregar borde exterior decorativo
        self.fig.add_shape(
            type="rect",
            xref="paper",
            yref="paper",
            x0=0,
            y0=0,
            x1=1,
            y1=1,
            line={
                'color': self.colors['primary'],
                'width': 3
            },
            fillcolor=None
        )

    def _draw_main_v_line(self):
        """Dibuja la l√≠nea principal en forma de V."""
        x_coords = [phase.coordinates[0] for phase in self.left_phases]
        y_coords = [phase.coordinates[1] for phase in self.left_phases]

        # A√±adir fase central
        x_coords.append(self.central_phase.coordinates[0])
        y_coords.append(self.central_phase.coordinates[1])

        # A√±adir fases derechas
        x_coords.extend([phase.coordinates[0] for phase in self.right_phases])
        y_coords.extend([phase.coordinates[1] for phase in self.right_phases])

        # Dibujar l√≠nea principal
        self.fig.add_trace(go.Scatter(
            x=x_coords,
            y=y_coords,
            mode='lines',
            line={
                'color': self.colors['line'],
                'width': self.design['line_width']
            },
            hoverinfo='skip'
        ))

    def _draw_connection_lines(self):
        """Dibuja las l√≠neas de conexi√≥n entre fases correspondientes."""
        for i, left_phase in enumerate(self.left_phases):
            right_phase = self.right_phases[len(self.right_phases) - 1 - i]

            # Dibujar l√≠nea de correspondencia
            self.fig.add_trace(go.Scatter(
                x=[left_phase.coordinates[0], right_phase.coordinates[0]],
                y=[left_phase.coordinates[1], right_phase.coordinates[1]],
                mode='lines',
                line={
                    'color': self.colors['connection_line'],
                    'width': self.design['connection_line_width'],
                    'dash': 'dot'
                },
                hoverinfo='skip'
            ))

    def _draw_phase_box(self, phase: VPhase):
        """Dibuja una fase individual con dise√±o mejorado."""
        # Calcular coordenadas del cuadro
        x0 = phase.coordinates[0] - phase.width/2
        y0 = phase.coordinates[1] - phase.height/2
        x1 = phase.coordinates[0] + phase.width/2
        y1 = phase.coordinates[1] + phase.height/2

        # Dibujar rect√°ngulo principal con borde suave
        self.fig.add_shape(
            type="rect",
            x0=x0,
            y0=y0,
            x1=x1,
            y1=y1,
            fillcolor=phase.color,
            line={
                'color': 'white',
                'width': self.design['phase_border_width']
            },
            layer='below'
        )

        # Texto del t√≠tulo de la fase
        self.fig.add_trace(go.Scatter(
            x=[phase.coordinates[0]],
            y=[phase.coordinates[1] + phase.height/5],  # Ligeramente arriba del centro
            mode='text',
            text=f"<b>{phase.name}</b>",
            textposition='middle center',
            textfont={
                'size': phase.title_font_size,
                'color': phase.text_color,
                'family': self.colors['font_family']
            },
            hoverinfo='text',
            hovertext=phase.hover_text if phase.hover_text else f"<b>{phase.name}</b><br>{phase.description}"
        ))

        # Texto de descripci√≥n dentro del cuadro
        self.fig.add_trace(go.Scatter(
            x=[phase.coordinates[0]],
            y=[phase.coordinates[1] - phase.height/5],  # Ligeramente abajo del centro
            mode='text',
            text=phase.description,
            textposition='middle center',
            textfont={
                'size': phase.desc_font_size,
                'color': phase.text_color,
                'family': self.colors['font_family']
            },
            hoverinfo='skip'
        ))

    def _add_section_labels(self):
        """A√±ade etiquetas para las secciones de Desarrollo y Validaci√≥n."""
        # Etiqueta de desarrollo
        self.fig.add_annotation(
            x=3,
            y=7,
            text="<b>DESARROLLO</b>",
            showarrow=False,
            font={
                'size': self.design['annotation_size'],
                'color': self.colors['primary'],
                'family': self.colors['font_family']
            }
        )

        # Etiqueta de validaci√≥n
        self.fig.add_annotation(
            x=11,
            y=7,
            text="<b>VALIDACI√ìN</b>",
            showarrow=False,
            font={
                'size': self.design['annotation_size'],
                'color': self.colors['primary'],
                'family': self.colors['font_family']
            }
        )

        # Flechas de direcci√≥n
        self.fig.add_annotation(
            x=7,
            y=7,
            text="",
            showarrow=True,
            arrowhead=2,
            arrowsize=1.5,
            arrowwidth=2,
            arrowcolor=self.colors['primary'],
            ax=-120,
            ay=0
        )

        self.fig.add_annotation(
            x=7,
            y=7,
            text="",
            showarrow=True,
            arrowhead=2,
            arrowsize=1.5,
            arrowwidth=2,
            arrowcolor=self.colors['primary'],
            ax=120,
            ay=0
        )

    def generate(self):
        """Genera el diagrama completo."""
        # Dibujar l√≠neas
        self._draw_main_v_line()
        self._draw_connection_lines()

        # Dibujar fases
        for phase in self.left_phases:
            self._draw_phase_box(phase)

        self._draw_phase_box(self.central_phase)

        for phase in self.right_phases:
            self._draw_phase_box(phase)

        # A√±adir etiquetas y flechas
        self._add_section_labels()

        return self.fig

    def show(self):
        """Muestra el diagrama interactivo."""
        self.generate().show()

    def save(self, filename: str, scale: int = 2, format: str = "png"):
        """
        Guarda el diagrama como imagen.

        Args:
            filename: Nombre del archivo sin extensi√≥n
            scale: Factor de escala para la resoluci√≥n (1-3)
            format: Formato de imagen (png, jpeg, svg, pdf)
        """
        full_filename = f"{filename}.{format}" if "." not in filename else filename
        self.generate().write_image(full_filename, scale=scale)

    def get_html(self):
        """Retorna el HTML para incrustar el diagrama en una p√°gina web."""
        return self.generate().to_html(include_plotlyjs=True, full_html=True)


# Ejemplo de uso con diferentes temas
if __name__ == "__main__":
    # Tema corporativo (por defecto)
    v_model_corporate = EnhancedVModelDiagram(
        title="MODELO EN V",
        subtitle="Metodolog√≠a de Desarrollo y Validaci√≥n",
        company="EUROLOGISTIC",
        theme="corporate"
    )
    v_model_corporate.show()

    # Tema moderno
    v_model_modern = EnhancedVModelDiagram(
        title="MODELO EN V",
        subtitle="Enfoque de Desarrollo de Software",
        company="EUROLOGISTIC",
        theme="modern"
    )
    # v_model_modern.show()

    # Tema cl√°sico
    v_model_classic = EnhancedVModelDiagram(
        title="MODELO EN V",
        subtitle="Proceso de Desarrollo de Software",
        company="EUROLOGISTIC",
        theme="classic"
    )
    # v_model_classic.show()

    # Guardar como imagen de alta calidad
    # v_model_corporate.save("modelo_v_corporativo_hq", scale=3)
```
